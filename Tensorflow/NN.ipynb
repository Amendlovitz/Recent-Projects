```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
f_mnist =  keras.datasets.fashion_mnist

(train_imgs, train_labs), (test_imgs, test_labs) = f_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
plt.figure()
plt.imshow(train_imgs[3])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_imgs = train_imgs / 255.0
test_imgs = test_imgs / 255.0
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape = (28, 28)), keras.layers.Dense(128, activation = 'relu'),
keras.layers.Dense(10, activation = 'softmax')])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

model.fit(train_imgs, train_labs, epochs = 8)
```

```python
test_loss, test_acc = model.evaluate(test_imgs, test_labs, verbose = 1)

print('test acc: ', test_acc, ' loss:', test_loss)
```

```python
predictions = model.predict(test_imgs)
print(class_names[np.argmax(predictions[5])])
```
