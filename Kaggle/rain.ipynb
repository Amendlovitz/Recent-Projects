```python
import pandas as pd
import numpy as np
import tensorflow as tf
from IPython.display import clear_output
```

```python
data = pd.read_csv("./Data/weatherAUS.csv")
data = data.dropna(subset=['RainTomorrow'])
data.RainTomorrow.replace(('Yes', 'No'), (0, 1), inplace=True)
```

```python
data.pop('Date')
data.head(10)
```

```python
for k in data:
    if data[k].dtypes == np.float64:
        data[k] -= np.mean(data[k])
        data[k] /= np.std(data[k])
data.head()
```

```python
for k in data:
    if data[k].isnull().values.any():
        if data[k].dtypes == np.float64:
            data[k].fillna(value=data[k].mean(), inplace=True)
        else:
            data[k].fillna(value=data[k].mode().iloc[0], inplace=True)
```

```python
shuffle_data = data.sample(frac=1)
train_size = int(0.7 * len(data))
train_set = shuffle_data[:train_size]
test_set = shuffle_data[train_size:]
train_y = train_set.pop("RainTomorrow")
test_y = test_set.pop("RainTomorrow")
```

```python
feature_columns = []
for k in train_set:
    if train_set[k].dtypes != np.float64:
        vocab = train_set[k].unique()
        feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(k, vocab))

    else:
        feature_columns.append(tf.feature_column.numeric_column(k, dtype=tf.float32))
```

```python
def make_input_fun(df, labels, epochs, shuffle=True, batch_size=32):
    def input_fun():
        ds = tf.data.Dataset.from_tensor_slices((dict(df), labels))

        if shuffle:
            ds = ds.shuffle(1000)
        
        ds = ds.batch(batch_size).repeat(epochs)
        return ds

    
    return input_fun


```

```python
train_input_fun = make_input_fun(train_set, train_y, 2)
test_input_fun = make_input_fun(test_set, test_y, 1, shuffle=False)
lin_est = tf.estimator.LinearClassifier(feature_columns=feature_columns)
lin_est.train(train_input_fun)
clear_output()
```

```python
res = lin_est.evaluate(test_input_fun)
clear_output()
```

```python
from tensorflow import keras
```

```python
for k in train_set:
    if train_set[k].dtypes != np.float64:
        train_set[k] = pd.Categorical(train_set[k])
        train_set[k] = train_set[k].cat.codes
clear_output()
```

```python
for k in test_set:
    if test_set[k].dtypes != np.float64:
        test_set[k] = pd.Categorical(test_set[k])
        test_set[k] = test_set[k].cat.codes
clear_output()
```

```python
NNmodel = keras.Sequential([keras.layers.Dense(21, input_dim=len(train_set.columns), activation='relu'),
                            keras.layers.Dense(10, activation='relu', use_bias=True),
                            keras.layers.Dense(1, activation='sigmoid')])
NNmodel.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics='accuracy')
```

```python
NNmodel.summary()
```

```python
NNmodel.fit(train_set, train_y, epochs=10)
clear_output()
```

```python
test_loss, test_acc = NNmodel.evaluate(test_set,  test_y, verbose = 1)

print('test acc: ', test_acc, ' loss:', test_loss)
```
